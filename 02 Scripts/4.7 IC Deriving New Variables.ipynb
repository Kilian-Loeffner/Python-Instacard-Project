### 2. Create customized column 

```python
# Create df with only 1 mio rows 

df = ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Create custom function to categorize products by price 

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply custom function to the dataframe in a new column 'price_range'

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Count values in price_range column

df['price_range'].value_counts(dropna = False)
```

### 3. Create customized column, using "loc()" function

```python
# Create conditions with the loc() function instead of IF statement 
# High Range

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Mid Range 
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 

# explaining: if = df.loc[(df['prices'] <= 15) & (df['prices'] > 5),
#             then = 'price_range_loc'] = 'Mid-range product'
```

```python
# Low Range

df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Count values in price_range_loc column

df['price_range_loc'].value_counts(dropna = False)
```
