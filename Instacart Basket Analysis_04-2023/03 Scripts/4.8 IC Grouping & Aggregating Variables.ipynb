## 2. Grouping and Aggregating data 

```python
# Group data with groupby() by "product_names"

df.groupby('product_name')
```

```python
# Group data by "department_id" and aggregate MEAN values from "order_number" column

df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Group data and get the mean value

df.groupby('department_id')['order_number'].mean()
```

```python
# Perform multiple aggregations

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## 3. Aggregating Data with "transform()"

```python
#create "max_order" column, group "user_id", look for max value in "order_number" within the groups

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## 4. Create Flag column 

```python
# Assigning values to "loyalty_flag" column 

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

## Step 5-7/ Task 4.8

```python
# 5. The marketing team at Instacart wants to know whether thereâ€™s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What youâ€™re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

# Look for average value in "price" column, grouped by customer type

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
# Count values in "price_range" column, grouped by customer type

ords_prods_merge.groupby('loyalty_flag')['price_range'].value_counts()
```

```python
# 6. he team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying.

# Create "avg_price" column, group by "user_id", look for mean price in the groups

ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output

ords_prods_merge.head(60)
```

```python
# Conditions for "spending_flag"

ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check output

ords_prods_merge[['user_id', 'avg_price', 'spending_flag']].head(200)
```

```python
# 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column.

# Create "avg_days_since_last_order" column, group by "user_id", look for median value from "days_since_last_order" column
ords_prods_merge['mean_days_since_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'mean_days_since_last_order']].head(50)
```

```python
# Conditions for order_frequenzy_flag column 

ords_prods_merge.loc[ords_prods_merge['mean_days_since_last_order'] > 20, 'order_frequenzy_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['mean_days_since_last_order'] > 10) & (ords_prods_merge['mean_days_since_last_order'] <= 20), 'order_frequenzy_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_days_since_last_order'] <= 10, 'order_frequenzy_flag'] = 'Frequent customer'
```

```python
# Check output 

ords_prods_merge[['user_id', 'mean_days_since_last_order', 'order_frequenzy_flag']].head(200)
```
