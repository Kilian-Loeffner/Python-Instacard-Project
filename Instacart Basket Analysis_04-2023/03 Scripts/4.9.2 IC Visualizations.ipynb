```python
# Create a bar chart, "sort_index" will sort data in order of the index

df_combined['orders_day_of_week'].value_counts().sort_index().plot.bar(color= 'black')
```

## 5. Sampling Data 

```python
# Seed funtion ensures the reproductivity, by making sure the random values are always the same

np.random.seed(4)
```

```python
# Create a list of random boolean values (len(ords_prods_merged)) sets the number of values to the number of rows in the dataframe.

dev = np.random.rand(len(df_combined)) <= 0.7
```

```python
dev
```

```python
# Create 70% sample

big = df_combined[dev]
```

```python
# Create 30% sample

small = df_combined[~dev]
```

```python
len(df_combined)
```

```python
len(big) + len(small)
```

```python
# Create subset only with columns necessary for the line chart

df_2 = small[['orders_day_of_week','prices']]
```

```python
df_2.describe().astype('int')
```

## 6. Line chart
